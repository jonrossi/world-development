# World Development Data Analysis #

This is a recreation of Hans Rosling's 2007 world development data visualization. It shows a significant, positive correlation between GDP per capita of a country and the average life expectancy of members of its population.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

**Read in the data set.**

```python
data = pd.read_csv('world_dev.csv')

gdp_cap = data['gdp_per_cap']
life_exp = data['life_expect']
pop = data['population']
col = data['color']
```

**Create a scatter plot** of the data by GDP per capita versus life expectancy. Color the data points by geographical region and size them based on the population of the country.

* Africa = blue
* Americas = yellow
* Asia = red
* Europe = green
* Oceania = black


```python
# Creat the scatter plot
plt.scatter(gdp_cap, life_exp, s=pop*2, c=col, alpha=0.6)
plt.xscale('log')

# Tick values and labels
tick_val = [1000, 10000, 100000]
tick_lab = ['1k', '10k', '100k']

# Adapt the ticks on the x-axis
plt.xticks(tick_val, tick_lab)

# Add axis labels
plt.xlabel('GDP per Capita (in USD)')
plt.ylabel('Life Expectancy (in years)')
plt.title('\n' + 'World Development in 2007' + '\n')

# Label India, China, and countries at the extreme ends of the graph
plt.text(1550, 69.5, 'India')
plt.text(5100, 78.25, 'China')
plt.text(132, 48, 'Congo, DR')
plt.text(44000, 73.7, 'USA')
plt.text(31000, 85.2, 'Japan')

# Add grid lines
plt.grid(True)

# Set the plot size
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 10
fig_size[1] = 6
plt.rcParams["figure.figsize"] = fig_size

# After customizing, display the plot
plt.show()
```
